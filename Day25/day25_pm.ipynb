# **PART A — Concept Application**

## **Problem Identification — Supervised / Unsupervised / Reinforcement**

**A system predicts whether an email is spam or not using past labeled emails.**

> **Supervised Learning — Classification.** The dataset has labeled examples (spam / not spam). The model learns to map input features (email content, sender, subject) to a discrete class label.

---

**A retail company groups customers based on purchasing behavior without any predefined labels.**

> **Unsupervised Learning — Clustering.** There are no target labels. The algorithm (e.g., k-means) discovers natural groupings in customer data entirely from patterns in the features.

---

**A robot learns to walk by trying different movements and receiving rewards when it moves correctly.**

> **Reinforcement Learning.** The robot (agent) takes actions in an environment, receives a reward or penalty signal, and iteratively learns a policy that maximises cumulative reward — no labeled dataset is involved.

---

**A model predicts house prices based on features like area, location, and number of rooms.**

> **Supervised Learning — Regression.** The target variable (price) is continuous. Labeled training data is used, and the model learns to output a number on a continuous scale.

---

**An e-commerce platform recommends similar products based on user browsing history.**

> **Unsupervised Learning (collaborative/content-based filtering).** There are no explicit outcome labels. Similarity is computed from patterns in browsing behavior using techniques like matrix factorisation or embeddings.

## **Regression vs Classification**

The single deciding factor is the **type of the target variable**:

| Target Variable | Problem Type | Example |
|---|---|---|
| Continuous number (price, salary, temperature) | **Regression** | Predict house price ₹42,00,000 |
| Discrete category / label | **Classification** | Predict if email is spam (Yes/No) |

**Justification:** Regression produces a value you could compute a mean for. Classification produces a label you would count. An important nuance — Logistic Regression, despite its name, is a **classification** algorithm: it outputs a probability which is then thresholded to a class label.

# **PART B — Stretch Problem**

## **Regression Example — Electricity Bill Prediction**

**Use case:** A power utility company wants to estimate a household's monthly electricity bill so they can send early alerts when projected consumption is too high.

**Input (features):** house area in sq ft, number of appliances, average daily temperature (°C), number of residents

**Output (target):** monthly electricity bill in ₹ — a continuous number

**Why regression:** The output can take any positive real value (₹800, ₹2,347.50, ₹6,800) — there are no discrete classes. A model must predict a number on a continuous scale.

```python
reg_data = {
    "area_sqft":      [800, 1200, 1500, 600, 2000, 900, 1100, 1800],
    "num_appliances": [3, 5, 7, 2, 9, 4, 5, 8],
    "avg_temp_c":     [28, 32, 30, 25, 35, 27, 29, 33],
    "num_residents":  [2, 4, 5, 1, 6, 3, 3, 5],
    "monthly_bill":   [1800, 3200, 4500, 1100, 6800, 2100, 2900, 5600]  # target (₹)
}

reg_df = pd.DataFrame(reg_data)

X_reg = reg_df.drop(columns=["monthly_bill"])
y_reg = reg_df["monthly_bill"]

print("X_reg:")
print(X_reg)
print()
print("y_reg (monthly_bill ₹):", y_reg.values)
```

## **Classification Example — Diabetes Prediction**

**Use case:** A hospital screening system flags patients at high risk of diabetes so doctors can prioritise follow-up tests.

**Input (features):** glucose level, BMI, age, blood pressure, insulin level

**Output (target):** 0 = no diabetes, 1 = diabetes — a discrete class label

**Why classification:** The outcome is one of a fixed set of categories (binary here). Even if the model internally computes a probability (e.g., 0.82), the final prediction is a label — diabetic or not — which makes it a classification problem.

```python
clf_data = {
    "glucose":        [85, 168, 140, 110, 189, 95, 155, 72],
    "bmi":            [22.3, 33.6, 30.1, 25.4, 38.2, 21.1, 29.7, 19.8],
    "age":            [22, 50, 44, 30, 60, 25, 47, 20],
    "blood_pressure": [70, 92, 88, 74, 100, 68, 84, 65],
    "insulin":        [85, 200, 160, 95, 230, 80, 175, 70],
    "diabetes":       [0, 1, 1, 0, 1, 0, 1, 0]  # target
}

clf_df = pd.DataFrame(clf_data)

X_clf = clf_df.drop(columns=["diabetes"])
y_clf = clf_df["diabetes"]

print("X_clf:")
print(X_clf)
print()
print("y_clf (diabetes):", y_clf.values)
```

## **Q1 — What are the types of Machine Learning?**

**1. Supervised Learning** — model is trained on labeled data (input + known output)
- **Regression** — predict a continuous value: house price, temperature, salary
- **Classification** — predict a discrete label: spam/not spam, disease/no disease
- Algorithms: Linear Regression, Logistic Regression, Decision Trees, SVM, Neural Networks

**2. Unsupervised Learning** — model finds patterns in unlabeled data
- **Clustering** — group similar data points: customer segmentation (k-means, DBSCAN)
- **Dimensionality Reduction** — compress features while preserving information: PCA, t-SNE
- **Association Rules** — find co-occurring patterns: market basket analysis (Apriori)

**3. Reinforcement Learning** — an agent learns by interacting with an environment and receiving reward/penalty signals
- No dataset required — experience is generated through trial and error
- Used in: games (AlphaGo), robotics, self-driving cars, stock trading bots

**4. Semi-supervised Learning** — small labeled dataset + large unlabeled dataset
- Useful when labeling is expensive: medical image annotation, speech recognition

**5. Self-supervised Learning** — labels are generated automatically from the data itself
- Foundation of modern LLMs: GPT predicts the next word using the sentence itself as the label

## **Q3 — Difference between Regression and Classification**

| | **Regression** | **Classification** |
|---|---|---|
| **Output** | Continuous number | Discrete class label |
| **Example output** | ₹42,500 salary | "Yes" (churn) / "No" |
| **Loss function** | MSE, MAE, RMSE | Cross-entropy, Log loss |
| **Evaluation metrics** | R², RMSE, MAE | Accuracy, F1-score, AUC-ROC |
| **Common algorithms** | Linear Regression, Ridge, SVR | Logistic Regression, Decision Tree, Random Forest, SVM |

**Key rule:** Look at the target variable.
- If it is a number you would compute a mean for → **Regression**
- If it is a label or category you would count → **Classification**

**Common confusion:** Logistic Regression is a classification algorithm despite the word *regression* in the name — it predicts a probability score (0.0–1.0) which is then converted to a class label using a threshold (typically 0.5).
