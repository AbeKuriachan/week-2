## Part A

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson, binom, norm, beta
```

```python
lam = 200
prob = 1 - poisson.cdf(220, lam)

prob
```

```python
n = 50
p = 0.02

prob = binom.pmf(3, n, p)
prob
```

```python
mu = 45
sigma = 8

p1 = 1 - norm.cdf(60, mu, sigma)
p2 = norm.cdf(50, mu, sigma) - norm.cdf(40, mu, sigma)

p1, p2
```

```python
lam = 10/60 * 6

prob = poisson.pmf(0, lam)
prob
```

```python
scores = np.random.normal(70, 10, (1000, 35))
means = scores.mean(axis=1)

plt.hist(means, bins=30, density=True)
plt.title("Distribution of Class Average (CLT)")
plt.show()
```

## PartB

```python
x = np.linspace(0,1,200)

plt.plot(x, beta.pdf(x,2,5), label="Beta(2,5)")
plt.plot(x, beta.pdf(x,5,5), label="Beta(5,5)")
plt.plot(x, beta.pdf(x,0.5,0.5), label="Beta(0.5,0.5)")

plt.legend()
plt.title("Beta Distribution Shapes")
plt.show()
```

```python
plt.plot(x, beta.pdf(x,1,1), label="Prior Beta(1,1)")
plt.plot(x, beta.pdf(x,8,4), label="Posterior Beta(8,4)")

plt.legend()
plt.title("Bayesian Update for Coin Bias")
plt.show()
```
