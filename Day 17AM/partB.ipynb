```python
import numpy as np
import time

N = 1_000_000

# Legacy
t0 = time.perf_counter()
np.random.seed(42)
_ = np.random.normal(0, 1, N)
legacy_ms = (time.perf_counter() - t0) * 1000

# New API
t0 = time.perf_counter()
rng = np.random.default_rng(42)
_ = rng.normal(0, 1, N)
new_ms = (time.perf_counter() - t0) * 1000

print(f"Legacy np.random.seed():  {legacy_ms:.1f} ms")
print(f"New    default_rng():     {new_ms:.1f} ms")
```

B.3 — What is np.random.Generator?
np.random.Generator (introduced in NumPy 1.17) is a class-based replacement for the
legacy module-level random functions. It wraps a BitGenerator (e.g., PCG64) and
exposes all the same distributions (normal, integers, etc.) as instance methods.
The main motivation was reproducibility and thread safety: the old np.random.seed()
set global state shared across the entire process, making parallel or multi-threaded
code fragile. Each Generator instance owns its own independent state, so you can
create one per thread or per experiment without interference.
It also allows swapping the underlying algorithm — you can pass np.random.MT19937()
to get the legacy Mersenne Twister, or np.random.PCG64() (the default) for better
statistical properties and speed.
