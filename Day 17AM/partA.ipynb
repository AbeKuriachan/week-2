```python
import numpy as np

rng = np.random.default_rng(2026)
grades = rng.integers(0, 101, size=(200, 8))

per_student_avg = grades.mean(axis=1)
per_course_avg  = grades.mean(axis=0)

print("Per-course averages (before curve):", np.round(per_course_avg, 1))
# → [47.8 56.0 50.7 53.4 46.9 51.3 48.8 48.9]

# Task 3 — Apply curve to courses with avg < 50, cap at 100
needs_curve = per_course_avg < 50
print("Courses getting curve:", np.where(needs_curve)[0].tolist())
# → [0, 4, 6, 7]

curved = grades.astype(float).copy()
curved[:, needs_curve] = np.clip(curved[:, needs_curve] + 10, 0, 100)

# Task 4 — Letter grades via np.where chain
letter_grades = np.where(curved >= 90, 'A',
                np.where(curved >= 80, 'B',
                np.where(curved >= 70, 'C',
                np.where(curved >= 60, 'D', 'F'))))

# Task 5 — Top 10 students by overall average
top10 = np.argsort(per_student_avg)[::-1][:10]
print("Top 10 student indices:", top10.tolist())
# → [84, 18, 34, 194, 164, 71, 30, 174, 195, 167]

# Task 6 — Students passing all courses (>= 60 in every course)
passed_all = np.all(grades >= 60, axis=1)
print(f"Students passing all courses: {passed_all.sum()} out of 200")
# → 0 out of 200  (seed=2026 produces harsh distribution)

# Overall stats
print(f"Overall mean: {grades.mean():.1f}, std: {grades.std():.1f}, "
      f"min: {grades.min()}, max: {grades.max()}")
```
