```python
import pandas as pd
df = pd.read_csv("eda_assignment_data.csv")
```

```python
import plotly.express as px
```

```python
fig = px.histogram(df, x="purchase_amount", nbins=40)
fig.write_html("plotly_sales_distribution.html")
```

```python
fig = px.box(df, x="product_category", y="purchase_amount")
fig.write_html("plotly_category_comparison.html")
```

```python
fig = px.scatter(df, x="annual_income", y="purchase_amount",
                 color="product_category")
fig.write_html("plotly_correlation_scatter.html")
```
