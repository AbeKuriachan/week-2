```python
jan_data = {
    "product": ["Laptop", "Phone", "Tablet", "Headphones", "Monitor", "Phone", "Laptop"],
    "units_sold": [5, 12, 6, 10, 4, 8, 3],
    "price": [60000, 30000, 20000, 5000, 15000, 30000, 60000]
}

feb_data = {
    "product": ["Laptop", "Phone", "Tablet", "Headphones", "Monitor", "Phone", "Tablet"],
    "units_sold": [6, 10, 8, 9, 5, 7, 4],
    "price": [60000, 30000, 20000, 5000, 15000, 30000, 20000]
}

mar_data = {
    "product": ["Laptop", "Phone", "Tablet", "Headphones", "Monitor", "Phone", "Laptop"],
    "units_sold": [7, 9, 5, 12, 6, 10, 4],
    "price": [60000, 30000, 20000, 5000, 15000, 30000, 60000]
}

jan_df = pd.DataFrame(jan_data)
feb_df = pd.DataFrame(feb_data)
mar_df = pd.DataFrame(mar_data)

jan_df, feb_df, mar_df
```

## **Summary Comparison**

```python
summary_df = pd.DataFrame(
    index=["January", "February", "March"],
    columns=["Total Revenue", "Average Order Value", "Top Selling Product"]
)

month_map = {
    "January": jan_df,
    "February": feb_df,
    "March": mar_df
}

for month, df in month_map.items():

    df["revenue"] = df["price"] * df["units_sold"]

    total_revenue = df["revenue"].sum()
    total_orders = df["units_sold"].sum()
    avg_order_value = total_revenue / total_orders

    top_product = (
        df.groupby("product")["units_sold"]
        .sum()
        .idxmax()
    )

    summary_df.loc[month] = [total_revenue, avg_order_value, top_product]

summary_df
```

## Q1: Difference Between `.loc[]` and `.iloc[]`

### `.loc[]`
- Label-based indexing
- Uses **row/column names**
- Slice **end is inclusive**

Example:


```python
df = pd.DataFrame(
    {"A": [10, 20, 30, 40, 50]},
    index=[0, 1, 2, 3, 4]
)

df.loc[0:3]
```

The result includes rows **0 through 3** because `.loc` includes the end index.

### `.iloc[]`
- Position-based indexing
- Uses **integer positions**
- Slice **end is exclusive**

```python
df.iloc[0:3]
```

The result includes rows **0,1,2** because `.iloc` excludes the end position.

### csv_analyser


```python
def analyze_csv(filepath):

    df = pd.read_csv(filepath)

    print("===== FIRST 5 MINUTES CHECKLIST =====")

    print("\nShape:")
    print(df.shape)

    print("\nColumns:")
    print(df.columns.tolist())

    print("\nFirst 5 rows:")
    print(df.head())

    print("\nData types:")
    print(df.dtypes)

    print("\nMissing values:")
    print(df.isnull().sum())

    print("\nSummary statistics:")
    print(df.describe())

    num_rows, num_cols = df.shape

    numeric_cols = df.select_dtypes(include="number").columns.tolist()
    categorical_cols = df.select_dtypes(exclude="number").columns.tolist()

    null_counts = df.isnull().sum().to_dict()

    memory_mb = df.memory_usage(deep=True).sum() / (1024**2)

    return {
        "num_rows": num_rows,
        "num_cols": num_cols,
        "numeric_cols": numeric_cols,
        "categorical_cols": categorical_cols,
        "null_counts": null_counts,
        "memory_mb": memory_mb
    }
```

```python
report = analyze_csv("budget_products.csv")
print(report)
```

```python
import pandas as pd

df = pd.DataFrame({
    "name": ["Alice", "Bob", "Charlie"],
    "age": [25, 30, 35],
    "salary": [50000, 60000, 70000]
})

# Fix 1: Use & instead of 'and' for element-wise conditions
high_earners = df[(df["age"] > 25) & (df["salary"] > 55000)]

# Fix 2: Avoid chained indexing, use .loc for assignment
df.loc[0, "age"] = 26

# Fix 3: iloc slice end is exclusive, so use 0:3 to include rows 0,1,2
first_three = df.iloc[0:3]

print("High earners:")
print(high_earners)

print("\nUpdated DataFrame:")
print(df)

print(f"\nGot {len(first_three)} rows, expected 3")
```
