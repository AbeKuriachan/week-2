## **detect_issues()**

```python
def detect_issues(df: pd.DataFrame) -> dict:

    # replace hidden NaN markers before counting
    df_check = df.replace(["N/A", "", "  ", " ", "None", "null"], np.nan)

    total_rows = len(df_check)
    total_missing = int(df_check.isnull().sum().sum())
    missing_per_column = df_check.isnull().sum().to_dict()
    duplicate_count = int(df_check.duplicated().sum())

    wrong_types = {
        col: str(df_check[col].dtype)
        for col in ["age", "rating", "purchase_amount"]
        if df_check[col].dtype == object
    }

    invalid_values = {}
    age_numeric = pd.to_numeric(df_check["age"], errors="coerce")
    invalid_ages = int(((age_numeric < 0) | (age_numeric > 120)).sum())
    if invalid_ages:
        invalid_values["age"] = f"{invalid_ages} values outside 0-120"

    rating_numeric = pd.to_numeric(df_check["rating"], errors="coerce")
    invalid_ratings = int(((rating_numeric < 1) | (rating_numeric > 5)).sum())
    if invalid_ratings:
        invalid_values["rating"] = f"{invalid_ratings} values outside 1-5"

    return {
        "total_rows": total_rows,
        "total_missing": total_missing,
        "missing_per_column": missing_per_column,
        "duplicate_count": duplicate_count,
        "wrong_types": wrong_types,
        "invalid_values": invalid_values
    }

report_before = detect_issues(df_raw)
report_before
```

## **clean_data()**

```python
def clean_data(df: pd.DataFrame) -> pd.DataFrame:

    df = df.copy()

    # step 1: replace hidden NaN markers
    df.replace(["N/A", "", "None", "null"], np.nan, inplace=True)
    df.replace(r'^\s+$', np.nan, regex=True, inplace=True)

    # step 2: fix types
    # to_numeric preferred over astype — coerces bad values to NaN instead of raising
    df["age"] = pd.to_numeric(df["age"], errors="coerce")
    df["rating"] = pd.to_numeric(df["rating"], errors="coerce")
    df["purchase_amount"] = pd.to_numeric(df["purchase_amount"], errors="coerce")

    # step 3: drop invalid age values (impossible ages — unfixable)
    df = df[~((df["age"] < 0) | (df["age"] > 120))]

    # step 4: standardize text columns
    # strip whitespace + lowercase for consistent grouping
    for col in ["name", "city", "feedback"]:
        df[col] = df[col].str.strip().str.lower()

    # step 5: fill missing values
    # age: median — skewed distribution likely, robust to outliers
    df["age"] = df["age"].fillna(df["age"].median())

    # rating: median — ordinal scale, median preserves middle tendency
    df["rating"] = df["rating"].fillna(df["rating"].median())

    # purchase_amount: median — monetary values often right-skewed, median > mean
    df["purchase_amount"] = df["purchase_amount"].fillna(df["purchase_amount"].median())

    # feedback: fill with 'no feedback' — can't impute free text meaningfully
    df["feedback"] = df["feedback"].fillna("no feedback")

    # step 6: parse dates with mixed formats using dayfirst=True
    df["signup_date"] = pd.to_datetime(df["signup_date"], dayfirst=True, errors="coerce")

    # step 7: remove duplicates — keep first occurrence
    df.drop_duplicates(subset="customer_id", keep="first", inplace=True)

    df.reset_index(drop=True, inplace=True)

    return df

df_clean = clean_data(df_raw)
print(df_clean.shape)
df_clean.head(10)
```

```python
from scipy import stats

def profile_dataframe(df: pd.DataFrame) -> dict:

    profile = {}

    for col in df.columns:
        series = df[col]
        col_profile = {
            "dtype": str(series.dtype),
            "unique_count": int(series.nunique()),
            "null_count": int(series.isnull().sum()),
            "null_pct": round(series.isnull().mean() * 100, 2),
            "top_5_values": series.value_counts().head(5).to_dict()
        }

        if pd.api.types.is_numeric_dtype(series):
            col_profile.update({
                "min": float(series.min()),
                "max": float(series.max()),
                "mean": round(float(series.mean()), 2),
                "median": float(series.median()),
                "std": round(float(series.std()), 2),
                "skewness": round(float(stats.skew(series.dropna())), 2)
            })
            # flag outliers > 3 std from mean
            mean, std = series.mean(), series.std()
            outlier_count = int(((series - mean).abs() > 3 * std).sum())
            if outlier_count:
                col_profile["suspicious_outliers"] = outlier_count

        elif pd.api.types.is_string_dtype(series) or series.dtype == object:
            str_lengths = series.dropna().astype(str).str.len()
            col_profile.update({
                "avg_length": round(float(str_lengths.mean()), 2),
                "min_length": int(str_lengths.min()),
                "max_length": int(str_lengths.max())
            })

        # potential issue flags
        if col_profile["unique_count"] == 1:
            col_profile["issue"] = "single-value column (useless feature)"
        elif pd.api.types.is_string_dtype(series) and col_profile["unique_count"] > 0.9 * len(df):
            col_profile["issue"] = "high-cardinality string"

        profile[col] = col_profile

    # print formatted summary
    print(f"{'='*50}")
    print(f"DataFrame Profile: {len(df)} rows x {len(df.columns)} cols")
    print(f"{'='*50}")
    for col, info in profile.items():
        print(f"\n[{col}] dtype={info['dtype']} | nulls={info['null_count']} ({info['null_pct']}%) | unique={info['unique_count']}")
        if "mean" in info:
            print(f"  numeric → min={info['min']} max={info['max']} mean={info['mean']} median={info['median']} skew={info['skewness']}")
        if "avg_length" in info:
            print(f"  string  → avg_len={info['avg_length']} min_len={info['min_length']} max_len={info['max_length']}")
        if "suspicious_outliers" in info:
            print(f"  ⚠ {info['suspicious_outliers']} outlier(s) detected (>3 std)")
        if "issue" in info:
            print(f"  ⚠ {info['issue']}")

    return profile

profile = profile_dataframe(df_clean)
```

## Q1: Handling 40% Missing in 'income'

With 40% missing in a 1M-row dataset, dropping rows would lose 400K records — too much data loss and likely introduces bias if the missingness is not random (e.g., lower-income respondents skip this field).

**Decision process:**
1. Check if missing is random (MCAR) or systematic (MAR/MNAR) — plot missingness against other columns
2. If MCAR: median fill is safe (income is typically right-skewed, so median > mean)
3. If MAR (e.g., missingness correlated with age/city): use group-wise median — `df.groupby('city')['income'].transform('median')`
4. If MNAR (people intentionally skipped): flag as a separate category or use a model-based imputation
5. For ML use cases: add a binary `income_was_missing` indicator column before filling

**Drop vs Fill:** Drop only rows where income is critical to the analysis AND the row has multiple other missing values. Otherwise fill with median or group median.


## **Q2: standardize_column()**

```python
def standardize_column(series: pd.Series) -> pd.Series:
    return (
        series
        .str.strip()                          # remove leading/trailing whitespace
        .str.lower()                          # lowercase
        .str.replace(r'[^a-z0-9\s]', '', regex=True)  # remove special characters
        .str.replace(r'\s+', ' ', regex=True) # collapse multiple spaces
        .str.strip()                          # strip again after collapse
    )

test_series = pd.Series([' Hello World!! ', ' NEW YORK ', 'san--francisco', ' MUMBAI '])
standardize_column(test_series)
```
