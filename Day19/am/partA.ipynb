```python
import numpy as np

def mean(arr):
    total = 0
    for x in arr:
        total += x
    return total / len(arr)

def median(arr):
    arr = sorted(arr)
    n = len(arr)
    mid = n // 2
    if n % 2 == 0:
        return (arr[mid-1] + arr[mid]) / 2
    return arr[mid]

def variance(arr):
    m = mean(arr)
    total = 0
    for x in arr:
        total += (x - m) ** 2
    return total / len(arr)


# higher-order version
from functools import reduce

def mean_ho(arr):
    return reduce(lambda a,b: a+b, arr) / len(arr)

def variance_ho(arr):
    m = mean_ho(arr)
    return reduce(lambda acc,x: acc + (x-m)**2, arr, 0) / len(arr)


# student marks
marks = [55, 72, 88, 45, 67, 91, 60]

avg = mean(marks)

above_avg = list(filter(lambda x: x > avg, marks))

grades = ["A" if x>=80 else "B" if x>=60 else "C" for x in marks]


# std deviation
def std_dev(arr):
    return variance(arr) ** 0.5


# verification
print(std_dev(marks), np.std(marks))


# sample groups
A = [12,23,3,34,45,56,5]
B = [12,1,3,1,1,2,3,4,5,3,4]

print(mean(A), variance(A))
print(mean(B), variance(B))
print("diff:", mean(A)-mean(B))
```
