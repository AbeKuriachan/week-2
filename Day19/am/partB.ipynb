```python
import numpy as np
from scipy import stats

def two_sample_t_test(a, b, alpha=0.05):
    mean1, mean2 = np.mean(a), np.mean(b)
    var1, var2 = np.var(a, ddof=1), np.var(b, ddof=1)

    n1, n2 = len(a), len(b)

    t_stat = (mean1 - mean2) / np.sqrt(var1/n1 + var2/n2)

    df = n1 + n2 - 2

    p_value = 2 * (1 - stats.t.cdf(abs(t_stat), df))

    if p_value < alpha:
        return t_stat, p_value, "Reject H0"
    return t_stat, p_value, "Fail to reject H0"


A = [12,23,3,34,45,56,5]
B = [12,1,3,1,1,2,3,4,5,3,4]

print(two_sample_t_test(A, B))
```
